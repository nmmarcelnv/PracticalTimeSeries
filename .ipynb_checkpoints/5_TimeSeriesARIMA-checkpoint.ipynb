# AR + MA = ARMA

**Auto Regressive AR(p)** and **Moving Average MA(q)** models by themselves may not be that interesting. Most real world processes are more complex, so we need to build more interesting models. This section is about how to combine **AR** and **MA** models to build more practical models.

---
**First some key results:**   

- For MA process, invertibility is important because it guarantees one to one mapping with ACF and the autocorrelation coefficients $\beta$'s) 

- For an MA(q) process, we have invertibility when the roots of 

$$\theta(ðµ) = \theta_0 + \theta_1ðµ + \theta_2ðµ^2 + â‹¯ + \theta_qðµ^q$$ 
all lie outside the unit circle.

- Given an AR(p) process we have stationarity when the roots of 

$$\phi(ðµ) = 1 âˆ’ \phi_1ðµ âˆ’ \phi_2ðµ âˆ’ â‹¯ âˆ’ \phi_pðµ$$ 
all lie outside the unit circle.

- Given an autoregressive AR(p) process, we can find an equivalent moving average MA($q = \infty$) process using the convenience of our shift operator B. The process
then has autocorrelation 

$$\rho(k) = \frac{ \sum_{i=0}^{\infty}{ \theta_i\theta_{i+k} } } {\sum_{i=0}^{\infty}{ \theta_i\theta_{i} } } $$   

- It may be very difficult to calculate $\theta$'s but we can use the Yule-Walker Equations to find $\rho_k$. 

$$\rho_k = \phi_1\rho_{k-1} + ... + \phi_p\rho_{k-p}$$ 

---

## Mixed AR and MA

**AR(p):** 

$$X_t = Z_t + \phi_1 X_{t-1} + \phi_2 X_{t-2} + ... + \phi_{t-p-1} + \phi_{t-p} X_{t-p}$$

$$\phi (B)X_t = Z_t$$

**MA(q):**  

$$X_t = Z_t + \theta_1 Z_{t-1} + ... + \theta_{t-q-1} + \theta_2 Z_{t-q} $$

$$X_t = \theta (B)Z_t$$

**ARMA(p,q):** 

>We can define
$X_t$ = <font color='red'> noise </font> + <font color='blue'> AR Part </font> + <font color='green'> MA Part </font>   
$$\theta (B)Z_t = \phi (B)X_t$$

>**We can express ARMA as MA($\infty$)**
$$X_t = \psi(B)Z_t$$
where $\psi(B) = \frac{\theta (B)}{\phi (B)}$

>**We can express ARMA as AR($\infty$)**
$$\pi(B)X_t = Z_t$$
where $\pi(B) = \frac{\phi(B)}{\theta(B)}$

**Why do we need to express ARMA as AR or MA?**  
> Because we know how to get parameters and properties of MA and AR processes already

**Simulated Example**

$$X_t = 0.7X_{t-1} + Z_t + 0.2Z_{t-1} \\ 
\Rightarrow X_t - 0.7X_{t-1} = Z_t + 0.2Z_{t-1} \\ 
\Rightarrow (1 - 0.7B)X_t = (1 + 0.2B)Z_t  \\
\Rightarrow \phi(B)X_t = \theta(B)Z_t   \\
$$

>**Writing as ARMA as AR($\infty$)**

$$\pi(B) = \frac{\phi(B)}{\theta(B)} = \frac{1 - 0.7B}{1 + 0.2B}$$  

$$\pi(B) = (1 - 0.7B)(1 + 0.2 ðµ)^{-1} \\
\Rightarrow \pi(B) =  (1 - 0.7B)(1 - 0.2B + 0.2^2B^2 -0.2^3B^3 + ...)  \\
\Rightarrow \pi(B) =  1 - 0.9B + 0.18B^2 -0.036B^3 + ...  \\
\Rightarrow \pi(B) =  1 - \sum_{k=1}^{\infty}{\pi_iB^k}
$$ 

>In general the coefficients of the term $B^k$ is $$(-1)^k(\phi+\theta)\theta^{k-1}$$

>**Writing as ARMA as MA($\infty$)**
$$X_t = \psi(B)Z_t$$
where $$\psi(B) = \frac{1 + \theta B}{1 - \phi B} = (1 + \theta B)(1 - \phi B)^{-1} \\
\Rightarrow \psi(B) = \sum_{k=1}^{\infty}{\psi_iB^k}
$$

>In general the coefficients of the term $B^k$ is $$(\phi+\theta)\phi^{k-1}$$ 

>With ðœ™ = .7, ðœƒ = .2   
$\psi_1 = 0.9, \psi_2 = .63, \psi_3 = .441, \psi_4 = .3087, \psi_5 = .21609, \psi_6 = .151263$  etc.. 

>**And we can show that the autocorrelation coefficients are given by**  

$$\rho_k = \frac{(1 + \phi \theta)(\phi + \theta)}{1+\theta^2+2\theta \phi}$$ 

**In the code**

````r
set.seed(500) 
# Wessex, East Anglia, Mercia, and Northumbria.
data = arima.sim( list(order = c(1,0,1), ar =.7, ma=.2), n = 10000)
par(mfcol = c(3,1 ))
plot(data, main="ARMA(1,1) Time Series: phi1=.7, theta1=.2", xlim=c(0,400)) #first terms
acf(data, main="Autocorrelation of ARMA(1,1), phi1=.7, theta1=.2")
acf(data, type="partial", main="Partial Autocorrelation of ARMA(1,1), phi1=.7, theta1=.2")
````

**General properties**   

|  NA    | AR(p)| MA(q)| ARMA(p,q) |
|------|------|------|-------|
|  ACF | decay|cut off after lag q| decay|
|  PACF | cut off after lag p|decay| decay| 

The above properties can be use to identify the order of ARMA models for fitting and forecasting. However, it's not always straighforward as it's often a judgement call to decide if a PACF or ACF plot show a cut-off or decay structure. Let's look at an example.

#### Real Dataet Example

Let's use a real dataset **discoveries** from the **R datasets library**. 
```R
help(discoveries)
```
**Description**  
The numbers of â€œgreatâ€ inventions and scientific discoveries in each year from 1860 to 1959.

````r
options(repr.plot.width=8, repr.plot.height=4)
par(mfcol = c(1,2))
plot(discoveries, main='Number of Major Scientific Discoveries')
stripchart(discoveries, method='stack', offset=0.5, at=0.15, pch=19, 
           main='Number of Major Scientific Discoveries',
          xlab='Year', ylab = 'discoveries')
````

````r
par(mfcol = c(1,2))
acf(discoveries)
pacf(discoveries)
````

- From the above, it's hard to decide if the ACF cuts off after lag 3 or has a decay structure. 
- The PACF appears to have a cut off after lag 1, but lag 2 is also very close to significant. So which model should we go with? 

**Grid Search is the Answer**  
- Look at models with combinations p = 0, 1, 2, 3 and q = 0, 1, 2, 3
- Calculate the AIC for each model
- Pick the simple model that minimizes the AIC

````r
ps = c(0,1,2,3)
qs = c(0,1,2,3)
aic = matrix(0, nrow = 4, ncol = 4) 

for (i in 1:4){
    p = ps[i]
    for (j in 1:4){
        q = qs[j]
        params = arima( discoveries, order=c(p,0,q) ) 
        aic[i,j] = params$aic
        print(paste("AIC for ARMA",p, q, aic[i, j]))
    }
} 
````

The most promissing models according to AIC are:

- ARMA(1,1) with AIC = 440.197 
- ARMA(3,2) with AIC = 439.926

Now is it worth using a much more complex model ARMA(3,2) for such a small difference in AIC? That's a judgement call. I will go with the simple model.

````r
arima( discoveries, order=c(1,0,1) ) 
````
